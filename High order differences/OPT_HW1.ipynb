**High-Order Derivatives for Wind Acceleration**

---


We employed a forward difference approximation for the estimation of wind acceleration in the globe for the NOAA data set, this is:
$$a_{i} = \frac{v_{i+1}-v_i}{h}\,,$$
where $v_i$ is the wind speed at time $i$, and $a_i$ denotes an estimation of the wind acceleration. Note that, by using this formula, we couldn't estimate the acceleration of the last time (because there is no $v_{i+1}$). However, we could do that by using a different formula, for instance, a backward difference approximation could make the work for the last time:
$$a_{i} = \frac{v_{i}-v_{i-1}}{h}\,,$$

**Exercise 1**

Propose two finite difference approximations for $f'(a)$ as follows:


1.   in the first one, only forward differences can be employed (i.e., $f(a+k\cdot h)$ for $k>0$), and

**Answer.**

1. Forward differences $(k > 0)$

```
from IPython.display import Latex, Display
import sympy as sp
from sympy.parsing.sympy_parser import parse_expr
```

```
def showL(s,eqn):
  display(Latex('(Eq. '+str(eqn)+'). \,'+sp.latex(s)+'=0'));
```

```
fx = parse_expr('f(x)');
a, h = parse_expr('a, h');

eq1 = fx.subs(x,a+h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+h)
showL(eq1,1)

eq2 = fx.subs(x,a+2*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+2*h)
showL(eq2,2)

eq3 = eq2 - 16*eq1
showL(eq3,3)

eq4 = fx.subs(x,a+3*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+3*h)
showL(eq4,4)

eq5 = fx.subs(x,a+6*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+6*h)
showL(eq5,5)

eq6 = (1/16)*eq5 - eq4
showL(eq6,6)

eq7 = eq6 - (27/16)*eq3
showL(eq7,7)

eq8 = fx.subs(x,a+4*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+4*h)
showL(eq8,8)

eq9 = fx.subs(x,a+8*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+8*h)
showL(eq9,9)

eq10 = eq9 - 16*eq8
showL(eq10,10)

eq11 = eq10 - 64*eq3
showL(eq11,11)

eq12 = eq11 - (128/3)*eq7
showL(eq12,12)
```

$f'(a)=\frac{-f(a+8h)+2.66666666666667f(a+6h)+16f(a+4h)-42.6666666666667f(a+3h)-8.0f(a+2h)+128.0f(a+h)-95.0f(a)-O(h5)}{56.0h}$

2. Backward differences $(k < 0)$

```
fx = parse_expr('f(x)');
a, h = parse_expr('a, h');

eq1 = fx.subs(x,a-h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a-h)
showL(eq1,1)

eq2 = fx.subs(x,a-2*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a-2*h)
showL(eq2,2)

eq3 = eq2 - 16*eq1
showL(eq3,3)

eq4 = fx.subs(x,a-3*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a-3*h)
showL(eq4,4)

eq5 = fx.subs(x,a-6*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a-6*h)
showL(eq5,5)

eq6 = (1/16)*eq5 - eq4
showL(eq6,6)

eq7 = eq6 - (27/16)*eq3
showL(eq7,7)

eq8 = fx.subs(x,a-4*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a-4*h)
showL(eq8,8)

eq9 = fx.subs(x,a-8*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a-8*h)
showL(eq9,9)

eq10 = eq9 - 16*eq8
showL(eq10,10)

eq11 = eq10 - 64*eq3
showL(eq11,11)

eq12 = eq11 - (128/3)*eq7
showL(eq12,12)
```

$f'(a)=\frac{128.0f(a−h)+8.0f(a−2h)-42.6666666666667f(a−3h)+16f(a−4h)+2.66666666666667f(a−6h)-f(a−8h)-95.0f(a)}{-56.0h}$
