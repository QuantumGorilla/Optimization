**High-Order Derivatives for Wind Acceleration**

---


We employed a forward difference approximation for the estimation of wind acceleration in the globe for the NOAA data set, this is:
$$a_{i} = \frac{v_{i+1}-v_i}{h}\,,$$
where $v_i$ is the wind speed at time $i$, and $a_i$ denotes an estimation of the wind acceleration. Note that, by using this formula, we couldn't estimate the acceleration of the last time (because there is no $v_{i+1}$). However, we could do that by using a different formula, for instance, a backward difference approximation could make the work for the last time:
$$a_{i} = \frac{v_{i}-v_{i-1}}{h}\,,$$

1. Forward differences $(k > 0)$

```
from IPython.display import display, Latex
import sympy as sp
from sympy.parsing.sympy_parser import parse_expr
```

```
def showEquation(s,eqn):
  display(Latex('(Eq. '+str(eqn)+'). \,'+sp.latex(s)+'=0'));
```

```
x = parse_expr('x'); 
fx = parse_expr('f(x)');
a, h = parse_expr('a, h');

eq1 = fx.subs(x,a+h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+h)
showEquation(eq1,1)

eq2 = fx.subs(x,a+2*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+2*h)
showEquation(eq2,2)

eq3 = eq2 - 16*eq1
showEquation(eq3,3)
```

```
eq4 = fx.subs(x,a+3*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+3*h)
showEquation(eq4,4)

eq5 = fx.subs(x,a+6*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+6*h)
showEquation(eq5,5)

eq6 = (1/16)*eq5 - eq4
showEquation(eq6,6)

eq7 = eq6 - (27/16)*eq3
showEquation(eq7,7)
```

```
eq8 = fx.subs(x,a+4*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+4*h)
showEquation(eq8,8)

eq9 = fx.subs(x,a+8*h)-sp.series(fx,x=x,x0=a,n=5).subs(x,a+8*h)
showEquation(eq9,9)

eq10 = eq9 - 16*eq8
showEquation(eq10,10)

eq11 = eq10 - 64*eq3
showEquation(eq11,11)
```

```
eq12 = eq11 - (128/3)*eq7
showEquation(eq12,12)
```

```
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
from netCDF4 import Dataset
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import seaborn as sns
```

```
u_wnd = Dataset('/content/drive/My Drive/uwnd.2020.nc','r');
v_wnd = Dataset('/content/drive/My Drive/vwnd.2020.nc','r');
```

```
lon = u_wnd['lon'][:];
lat = v_wnd['lat'][:];
u = u_wnd['uwnd'];
v = v_wnd['vwnd'];
u = u[:,0,:,:];
v = v[:,0,:,:];
ws = np.sqrt(np.power(u,2)+np.power(v,2))
times, nlat, nlon = np.shape(ws);
ac = np.zeros((times-1,nlat,nlon));
n_ws=len(ws)
```

```
def fwd_diff(f8, f6, f4, f3, f2, f1, f0, h): #Approach from the right
  return (-f8+2.66666666666667*f6+16*f4-42.6666666666667*f3-8.0*f2+128.0*f1-95.0*f0)/(56.0*h)
```

```
def fbwd_diff(f8, f6, f4, f3, f2, f1, f0, h): #Approach from the left
  return (128.0*f1+8.0*f2-42.6666666666667*f3+16*f4+2.66666666666667*f6-f8-95.0*f0)/(-56.0*h)
```

```
for i in range (0,n_ws-8):
  ac[i,:,:] = fwd_diff(ws[i+8,:,:],ws[i+6,:,:],ws[i+4,:,:],ws[i+3,:,:],ws[i+2,:,:],ws[i+1,:,:],ws[i,:,:],0.25);
for i in range (n_ws-7,n_ws-1):
  ac[i,:,:] = fbwd_diff(ws[i-8,:,:],ws[i-6,:,:],ws[i-4,:,:],ws[i-3,:,:],ws[i-2,:,:],ws[i-1,:,:],ws[i,:,:],0.25);
```

```
ac_mean = np.mean(ac,0); #Average accelerations 
lonM,latM = np.meshgrid(lon, lat);
fig = plt.figure(figsize=(10,8));
m = Basemap(resolution='l',projection='robin',lat_0=lat[0],lon_0=lon[0]);
m.shadedrelief(scale=0.5);
m.drawcoastlines(linewidth=1);
m.drawcountries(linewidth=1);
m.drawparallels(np.arange(-90,90,20));
m.drawmeridians(np.arange(0,357.5,20));
m.contourf(lonM,latM,np.abs(ac_mean),latlon=True,cmap='jet',alpha=0.6);
cbar = m.colorbar(location='bottom');
cbar.set_label(label='W');
cbar.set_label(label='Wind Acceleration in $m/s^2$', size=25);
plt.show();
```
